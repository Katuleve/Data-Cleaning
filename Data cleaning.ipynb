```python
numeric_columns = df.select_dtypes(include=['number']).columns
```

```python
for column in numeric_columns:
    column_mean = df[column].mean()
    df[column].fillna(column_mean)
df
```

```python
df['GPA'] = pd.to_numeric(df['GPA'], errors='coerce')
df_GPA_mean = df['GPA'].mean()
df['GPA'].fillna(df_GPA_mean)
df['GPA'].unique()
```

```python
object_columns = df.select_dtypes(include=['object']).columns

df[object_columns]
```

```python
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
weight_mean = df['weight'].mean()
df['weight'].fillna(weight_mean)
df['weight'].unique()
```

```python
object_columns = df.select_dtypes(include=['object']).columns

for column in object_columns:
    unique_column_values = df[column].nunique()
    print(f' Column {column} has {unique_column_values} unique values')
```
